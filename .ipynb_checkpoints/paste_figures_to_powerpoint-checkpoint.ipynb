```python
from pptx import Presentation
from glob import glob
from PIL import Image
from pptx.util import Pt
```

```python
# Input
grid_width = 3
grid_height = 2
fnms = glob('img/tokyo*.JPG')
is_new_ppt =True
file_name = 'とうきょう.pptx'
text_pic_ratio = 20
font_size = 16
```

```python
# パワポの準備
if is_new_ppt:
    ppt = Presentation()
else:
    ppt = Presentation(file_name)
width = ppt.slide_width
height = ppt.slide_height
```

```python
#使用するスライドの種類
title_slide_layout = ppt.slide_layouts[0] #Title Slideの作成
bullet_slide_layout = ppt.slide_layouts[1] #Title and Contentの作成
blank_slide_layout = ppt.slide_layouts[6] #Blankの作成
```

```python
#スライドに画像貼り付け
slide_picture = ppt.slides.add_slide(blank_slide_layout)
count = 0
for h in range(grid_height):
    for w in range(grid_width):
        fnm = fnms[count]
        img = Image.open(fnm)
        #　横長画像→横幅で画像サイズ調整　縦長画像→立幅で画像サイズ調整
        aspect_ratio = img.width / img.height
        if aspect_ratio > width / height:
            slide_picture.shapes.add_picture(fnm,w/grid_width*width,h/grid_height*height+height/text_pic_ratio,width = width/2)
        else:
            slide_picture.shapes.add_picture(fnm,w/grid_width*width,h/grid_height*height+height/text_pic_ratio,height = height/2-height/text_pic_ratio)
        #　テキストボックス貼り付け         
        txBox = slide_picture.shapes.add_textbox(w/grid_width*width, h/grid_height*height, height=height/text_pic_ratio, width = width/grid_width)
        tB = txBox.text_frame
        tB.text = str(fnms[count]).split('\\')[-1].split('.')[0]
        tB.paragraphs[0].font.size = Pt(font_size)
        count += 1
        if len(fnms) <= count:
            break
ppt.save(file_name)
```
